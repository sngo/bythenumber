```python
#get unique name of entity
fileData.Entity.unique()
```

```python
viet = fileData.loc[fileData['Entity'] == 'Vietnam']
```

```python
viet.shape
```

```python
viet.head(3)
```

```python
vietshort = viet.drop(['Entity', 'Code'], axis=1)
```

```python
vietshort.head(3)
```

```python
columnNames = vietshort.columns.values.tolist()
```

```python
columnNames[0]
```

```python
shortColumnNames = list()
for name in columnNames:
    #print(i)
    if(name != 'Year'):
        tempSplit = name.split("-")
        shortColumnNames.append(tempSplit[1])
    else:
        shortColumnNames.append('Year')
```

```python
print(shortColumnNames)
```

```python
vietshort.columns = shortColumnNames
```

```python
vietshort.head(3)
```

```python
noindex = (vietshort.set_index('Year')).transpose()
#noindex.transpose()
```

```python
noindex.to_csv('vietnam-death-by-cancer-type.csv')
```
